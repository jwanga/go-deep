```python
%matplotlib inline
import math
import numpy as np
import matplotlib.pyplot as plt

#Generate a list of numbers from -5, 5 in .1 increments.
inputs = [x * .1 for x in range(-50, 51)]

#Define a bias value
global_bias = 1
```

```python
def linearActivationFunction(x):
    '''
    Linear activation functions do not change the value of the input
    They are typically used in the output layer of a network learning
    Numerical Values.
    '''
    y = x
    
    return y


outputs = [linearActivationFunction(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
def stepActivationFunction(x):
    '''
    Step activation functions output a zero if the input is below a certain 
    threshold and a 1 if the input is at or above that threshold.
    '''
    y = 1 if x >= .5 else 0


outputs = [stepActivationFunction(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
def sigmoidActivationFunction(x):
    '''
    Sigmoid activation functions output a only positive numbers within a very narow range.
    '''
    y = 1 / (1 +math.exp(-x))
    
    return y

outputs = [sigmoidActivationFunction(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
def HyperbolicTangentActivationFunction(x):
    '''
    Hyperbolic tangent activation functions output a very narow range between -1 and 1.
    '''
    y = math.tanh(x)
    
    return y

outputs = [HyperbolicTangentActivationFunction(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
def RectifiedLinearUnitActivationFunction(x):
    '''
    Rectified Linear Unit activation functions scale linealy above a minimum threshold.
    '''
    y = max(0, x)
    
    return y

outputs = [RectifiedLinearUnitActivationFunction(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
def SoftmaxActivationFunction(output_neurons):
    '''
    Softmax activation functions output a vector containing probabilities that each neuron in an output layer 
    is a member of the class represented by the Softmax function.
    '''
    #sum the exponent of each ouptut neuron value
    output_neuron_sum = sum([math.exp(x) for x in output_neurons])
    
    #Get probabilities by dividing the exponent of each output newron with the sum of the exponent of 
    #all output neurons.
    probabilities = [math.exp(x) / output_neuron_sum for x in output_neurons]
    
    return probabilities

output_neurons = [.9, .2, .4]
probabilities = SoftmaxActivationFunction(output_neurons)

print(output_neurons);
print(probabilities);
```

```python
def SigmoidActivationNeuron(weight, input, bias):
    '''
    A neuron that combines a sigmoid activation function, input value and a bias value.
    '''
    
    return sigmoidActivationFunction(weight * input + bias)
    
# plot the output for the neuron with different weights so we can obserse the effect on the output.
for i in [.5, 1, 1.5, 2]:
    plt.plot(inputs, [SigmoidActivationNeuron(i, x, 0) for x in inputs]);

plt.show();
```
