```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import functools
import itertools
import operator
import math
import neurons

#Generate a list of numbers from -5, 5 in .1 increments.
inputs = [x * .1 for x in range(-50, 51)]

#Define a bias value
global_bias = 1
```

```python
activation = neurons.LinearActivationFunction()
outputs = [activation.compute(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python


activation = neurons.StepActivationFunction()
outputs = [activation.compute(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
activation = neurons.SigmoidActivationFunction()
outputs = [activation.compute(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
activation = neurons.HyperbolicTangentActivationFunction()
outputs = [activation.compute(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
activation = neurons.RectifiedLinearUnitActivationFunction()
outputs = [activation.compute(x) for x in inputs]
line = plt.plot(inputs, outputs);
plt.show();
```

```python
activation = neurons.SoftmaxActivationFunction()
output_neurons = [.9, .2, .4]
probabilities = activation.compute(output_neurons)

print(output_neurons);
print(probabilities);
```

```python
# plot the output for the neuron with different weights so we can observe the effect on the output.
for i in [.5, 1, 1.5, 2]:
    neuron = neurons.StepActivationNeuron([i], 0)
    plt.plot(inputs, [neuron.compute([x]) for x in inputs]);

plt.show();

#plot the output for the neuron with different bias so we can observe the effect on the output.
for i in [.5, 1, 1.5, 2]:
    neuron = neurons.StepActivationNeuron([1], i)
    plt.plot(inputs, [neuron.compute([x]) for x in inputs]);

plt.show();
```

```python
# You can hand train simple networks to represent basic logical operators

and_neuron = neurons.StepActivationNeuron([1, 1], -1.5);
or_neuron = neurons.StepActivationNeuron([1, 1], -.5);
not_neuron = neurons.StepActivationNeuron([-1], .5);

pairs = [[1, 1], [1, 0], [0, 1], [0, 0]]

print('AND')
for pair in pairs:
    print('{0} | {1} | {2}'.format(pair[0], pair[1], and_neuron.compute([pair[0], pair[1]])))

print('\nOR')
for pair in pairs:
    print('{0} | {1} | {2}'.format(pair[0], pair[1], or_neuron.compute([pair[0], pair[1]])))

print('\nNOT')
for i in [1, 0]:
    print('{0} | {1}'.format(i, not_neuron.compute([i])))

print('\nXOR')
for pair in pairs:
    layer_1_0 = or_neuron.compute([pair[0], pair[1]])
    layer_1_1 = and_neuron.compute([pair[0], pair[1]])
    layer_2_0 = not_neuron.compute([layer_1_1])
    layer_3_0 = and_neuron.compute([layer_1_0, layer_2_0])
    print('{0} | {1} | {2}'.format(pair[0], pair[1], layer_3_0))
    
print('\nNOR')
for pair in pairs:
    layer_1_0 = or_neuron.compute([pair[0], pair[1]])
    layer_2_0 = not_neuron.compute([layer_1_0])
    print('{0} | {1} | {2}'.format(pair[0], pair[1], layer_2_0))
    
print('\nNAND')
for pair in pairs:
    layer_1_0 = and_neuron.compute([pair[0], pair[1]])
    layer_2_0 = not_neuron.compute([layer_1_0])
    print('{0} | {1} | {2}'.format(pair[0], pair[1], layer_2_0))
```
